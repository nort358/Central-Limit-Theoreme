```
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import random
import statistics
import seaborn as sns
from scipy.stats import norm
xvalues=[]
xvalues1=[]
xvalues2=[]
random.seed(0)

#random samples generating and calculating their means:
for y in range (1000):
  z=statistics.mean([random.random() for x in range(3)])
  z1=statistics.mean([random.random() for x in range(15)])
  z2=statistics.mean([random.random() for x in range(100)])
  xvalues.append(z)
  xvalues1.append(z1)
  xvalues2.append(z2)

mean=statistics.mean(xvalues)
stdev=statistics.stdev(xvalues)
mean1=statistics.mean(xvalues1)
stdev1=statistics.stdev(xvalues1)
mean2=statistics.mean(xvalues2)
stdev2=statistics.stdev(xvalues2)

#building 3 fixed fields
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey='all',sharex=True)
#
axes[0].set_title('Frequency Histogram for 3 sized samples')  
axes[1].set_title('Frequency Histogram for 15 sized samples')  
axes[2].set_title('Frequency Histogram for 100 sized samples')  

#plotting 3 histograms and their density
sns.distplot(xvalues,ax=axes[0],color='dodgerblue',label='avg= '+str(round(mean,3))+'\n stdev= '+str(round(stdev,3)), axlabel='Value')
sns.distplot(xvalues1,ax=axes[1],color='deeppink',label='avg= '+str(round(mean1,3))+'\n stdev= '+str(round(stdev1,3)),axlabel='Value')
sns.distplot(xvalues2,ax=axes[2],color='gold',label='avg= '+str(round(mean2,3))+'\n stdev= '+str(round(stdev2,3)),axlabel='Value')
for ax in axes:
  ax.legend(loc=0)
  ax.set_ylabel('Frequency')

#plotting normal distribution - the black one
sns.distplot(np.random.normal(0.5,(1/12)**(1/2),10000),ax=axes[0],hist=False, color='black')
sns.distplot(np.random.normal(0.5,(1/12)**(1/2),10000),ax=axes[1],hist=False, color='black')
sns.distplot(np.random.normal(0.5,(1/12)**(1/2),10000),ax=axes[2],hist=False, color='black')

#with n increase, there is a tendention to the actual average, distibutions are getting narrower
```
